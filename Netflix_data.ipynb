```python
# Split 'listed_in' (genres) into lists
df['listed_in_list'] = df['listed_in'].apply(lambda x: [i.strip() for i in x.split(',')] if pd.notnull(x) else [])

# Flag for multiple genres
df['multi_genre'] = df['listed_in_list'].apply(lambda x: len(x) > 1)

# Create content length feature (dummy: e.g., for Movies - 1, TV Shows - 10)
df['content_length'] = np.where(df['type'] == 'Movie', 1, 10)

# Handle inconsistent country names (example)
df['country'] = df['country'].replace({'United States of America': 'United States'})

# Outlier check for release years
df = df[(df['release_year'] >= 1920) & (df['release_year'] <= 2025)]
```

```python
# Extract year, month, and weekday from date_added
df['year_added'] = df['date_added'].dt.year
df['month_added'] = df['date_added'].dt.month
df['weekday_added'] = df['date_added'].dt.day_name()
```

```python
df.dropna(subset=['cast'], inplace=True) #drop the row ,where cast value missing
df['date_added'] = df['date_added'].str.strip() #clean space and convert to datetime 
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
```

```python
df=df.dropna(subset=['date_added'])   #Drop the row date_added with invalid or misising value 
df['director'].fillna('Unknown',inplace=True)  #fill the misisng value in director column with Unknowm
df['country'].fillna('Unknown',inplace=True)  # fill the missing value with unknown in country column
df['rating']=df['rating'].fillna(df['rating'].mode()[0])  #fill the rating missing value with mode value 
df['duration']=df['duration'].fillna(df['duration'].mode()[0]) #fill the duration misiing value with the mode value
```

```python
# Aggregation example: Number of titles by country
country_counts = df['country'].value_counts().head(10)
print("\nTop 10 Countries by Content:\n", country_counts)

# Final dataset preview
print("\nCleaned Netflix Dataset:")
print(df.head())
     
```

```python
sns.countplot(data=df, x='type', palette='Set2') #plotting the Content type Distribution Graph
plt.title("Content Type Distribution")
plt.show()
```

```python
# Plot 3: Number of titles added over time
plt.figure(figsize=(12, 6))
df['year_added'] = pd.to_datetime(df['date_added']).dt.year
yearly_counts = df['year_added'].value_counts().sort_index()
sns.lineplot(x=yearly_counts.index, y=yearly_counts.values, marker='o')
plt.title('Number of Titles Added Over Time')
plt.xlabel('Year Added')
plt.ylabel('Number of Titles')
plt.tight_layout()
plt.show()
```

```python
# Plot 4: Most common genres (from 'listed_in')
from collections import Counter

all_genres = df['listed_in'].dropna().str.split(', ')
flat_genres = [genre for sublist in all_genres for genre in sublist]
genre_counts = Counter(flat_genres).most_common(10)

genres, counts = zip(*genre_counts)
plt.figure(figsize=(10, 6))
sns.barplot(x=counts, y=genres, palette='coolwarm')
plt.title('Top 10 Most Common Genres on Netflix')
plt.xlabel('Number of Titles')
plt.ylabel('Genre')
plt.tight_layout()
plt.show()
```
